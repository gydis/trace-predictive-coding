```python
from sympy import *
```

```python
init_printing(use_unicode=True)
```

```python
mu0, mu1, mu2 = symbols(r'\mu_0 \mu_1 \mu_2')
th0, th1, th2 = symbols(r'\theta_0 \theta_1 \theta_2', )
x0, x1, x2 = symbols(r'x_0 x_1 x_2')
S_0, S_1, S_2 = symbols(r'\Sigma_0 \Sigma_1 \Sigma_2')
```

```python
F = S_1**-1 * (mu1 - th2*mu2)**2 + S_0**-1 * (mu0 - th1*mu1)**2 + S_2**-1 * (mu2 - th2*x2)**2
```

```python
diff(F, mu0)
```

```python
diff(F, mu1)
```

```python
diff(F, mu2)
```

```python
diff(F, th2)
```
